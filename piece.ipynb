## Measure dictionaries
All information about a single measure is stored in the dictionary p.measure which can be accessed as p:

```python
p[52]
```

The keys of the dictionary are the following:

Each measure has an 'events' dictionary whose keys are each event's offset from beat 1 in ticks. The corresponding value, i.e. the event, is itself a dictionary which looks like this:

BTW, only events from the upper system are saved. Events from secondary voices are saved only if they don't have the same onset as an event in the main voice. Also, pitches are not saved. The parser was built mainly for dealing with harmony labels.

```python
len(p[24]['events'])
```
